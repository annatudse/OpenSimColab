## Step 3: Plot GRFs and finally



```python
#@title Print some info
# first get some info from the c3d and print based on the following link and script
# from https://github.com/pyomeca/ezc3d/blob/dev/examples/python3/force_platform.py
# see the github of the dev. above for amazing work they do! 

import ezc3d

# you can change the sample file below based on what you downloaded
c3d = ezc3d.c3d("/content/Sample10/TYPE-2a.c3d", extract_forceplat_data=True)

print(f"Number of force platform = {len(c3d['data']['platform'])}")
print("")
print("Printing information and data for force platform 0")
print("")
pf0 = c3d["data"]["platform"][0]
# Units
print(f"Force unit = {pf0['unit_force']}")
print(f"Moment unit = {pf0['unit_moment']}")
print(f"Center of pressure unit = {pf0['unit_position']}")
print("")
# Position of pf
print(f"Position of origin = {pf0['origin']}")
print(f"Position of corners = \n{pf0['corners']}")
print("")
# Calibration matrix
print(f"Calibation matrix = \n{pf0['cal_matrix']}")
print("")
# Data at 3 different time
frames = [0, 10, 1000, -1]
print(f"Data (in global reference frame) at frames = {frames}")
print(f"Force = \n{pf0['force'][:, frames]}")
print(f"Moment = \n{pf0['moment'][:, frames]}")
print(f"Center of pressure = \n{pf0['center_of_pressure'][:, frames]}")
print(f"Moment at CoP = \n{pf0['Tz'][:, frames]}")
```

```python
#@title Plot GRFs
# plot GRFs in one of the force plates
import matplotlib.pyplot as plt
import numpy as np

a = pf0['force']
x1 = np.array(a[0,:]) 
x2 = np.array(a[1,:])
x3 = np.array(a[2,:]) #GRF vertical
plt.title("GRF graph")

plt.plot(x1)
plt.plot(x2)
plt.plot(x3)
plt.show()
```
